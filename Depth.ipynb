```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)
qc.x(0)
qc.measure([0], [0])

print(qc.depth())
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 1)

qc.x(0)
qc.barrier()
qc.cx(0, 2)
qc.cx(1, 2)
qc.ccx(0, 1, 2)
qc.barrier()

qc.measure(2, 0)
display(qc.draw('mpl'))
print(qc.depth())
```

```python
qc = QuantumCircuit(3)

qc.h(0)
qc.cx(0, 1)
qc.h(0)
qc.h(1)
qc.cx(0, 2)
display(qc.draw('mpl'))
print(qc.depth())
```

```python
qc = QuantumCircuit(4, 1)

qc.x(range(4))
qc.y(1)

qc.barrier()
qc.cx(0, 2)
qc.cx(1,3)
qc.cx(0, 2)
# qc.barrier()


qc.measure(2, 0)
print(qc.depth())
qc.draw('mpl')
```

```python
qc_ha = QuantumCircuit(5,2)
# encode inputs in qubits 0 and 1
qc_ha.x(range(5)) # For a=0, remove the this line. For a=1, leave it.
qc_ha.x(1) # For b=0, remove the this line. For b=1, leave it.
qc_ha.y(3)
qc_ha.barrier()
# use cnots to write the XOR of the inputs on qubit 2
qc_ha.cx(0,2)
qc_ha.cx(3,4)
# use ccx to write the AND of the inputs on qubit 3
qc_ha.ccx(0,1,3)
qc_ha.cx(2,4)
qc_ha.barrier()
# extract outputs
qc_ha.measure(2,0) # extract XOR value
qc_ha.measure(3,1) # extract AND value

display(qc_ha.draw('mpl'))
print('Depth of a circuit :', qc_ha.depth())
```
